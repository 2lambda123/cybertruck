This script randomly selects 100 images from each of the 10 classes in 'v1_cam1_no_split'. Will be used for fine-tuning yolov model.

```python
import os
import shutil
from glob import glob
import random

def export_some_imgs(path):
    
    dest_folder = '../../utils/datasets/distracted_driving'

    folders = glob(path + '/*')
    folders = [file for file in folders if not file.endswith('.csv')]
    for folder in folders:
        file_list = glob(folder +'/*jpg')
        print(f'The folder: {folder}')
        print(file_list)
        print(f'File Length: {len(file_list)}')
        name = folder.split('/')[-1].lower().replace(' ', '_')
        folder_subset = random.choices((file_list), k=100)

        for idx, img in enumerate(folder_subset):
            # orig_path = os.path.join(folder, img)
            dest_path = os.path.join(dest_folder, f'{name}_{idx}.jpg')
            print(img)
            # print(img)
            shutil.copy(img, dest_path)


```

```python
root = os.path.dirname(os.path.abspath(os.getcwd()))

export_some_imgs(path=root + '/v1_cam1_no_split')
```
